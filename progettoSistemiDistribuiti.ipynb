###Donwload del database

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
cdc_diabetes_health_indicators = fetch_ucirepo(id=891)

# data (as pandas dataframes)
X = cdc_diabetes_health_indicators.data.features
y = cdc_diabetes_health_indicators.data.targets
```

##Suddivisione del db per ciascun nodo

```python
X_split = np.array_split(X, 4)
print('Shape First Part: ')
print(X_split[0].shape)
print('Shape Second Part: ')
print(X_split[1].shape)
print('Shape Third Part: ')
print(X_split[2].shape)
print('Shape Fourt Part: ')
print(X_split[3].shape)

print('----------------------')

y_split = np.array_split(y, 4)
print('Shape First Part: ')
print(y_split[0].shape)
print('Shape Second Part: ')
print(y_split[1].shape)
print('Shape Third Part: ')
print(y_split[2].shape)
print('Shape Fourt Part: ')
print(y_split[3].shape)
```

####Suddivisione in train e test

```python
trainX_a, testX_a, trainY_a, testY_a = train_test_split(X_split[0], y_split[0], test_size=0.2)
trainX_b, testX_b, trainY_b, testY_b = train_test_split(X_split[1], y_split[1], test_size=0.2)
trainX_c, testX_c, trainY_c, testY_c = train_test_split(X_split[2], y_split[2], test_size=0.2)
trainX_d, testX_d, trainY_d, testY_d = train_test_split(X_split[3], y_split[3], test_size=0.2)
```

```python
model_a = Sequential()
model_a.add(Dense(3, activation='relu', input_shape=(trainX_a.shape[1], 1)))
model_a.add(Dropout(0.1))
model_a.add(Flatten())
model_a.add(Dense(1, activation='sigmoid'))
model_a.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_a.summary()

model_b = Sequential()
model_b.add(Dense(3, activation='relu', input_shape=(trainX_a.shape[1], 1)))
model_b.add(Dropout(0.1))
model_b.add(Flatten())
model_b.add(Dense(1, activation='sigmoid'))
model_b.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_b.summary()

model_c = Sequential()
model_c.add(Dense(3, activation='relu', input_shape=(trainX_a.shape[1], 1)))
model_c.add(Dropout(0.1))
model_c.add(Flatten())
model_c.add(Dense(1, activation='sigmoid'))
model_c.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_c.summary()

model_d = Sequential()
model_d.add(Dense(3, activation='relu', input_shape=(trainX_a.shape[1], 1)))
model_d.add(Dropout(0.1))
model_d.add(Flatten())
model_d.add(Dense(1, activation='sigmoid'))
model_d.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_d.summary()
```

```python
history_a = model_a.fit(trainX_a, trainY_a, epochs=3)
history_b = model_b.fit(trainX_b, trainY_b, epochs=3)
history_c = model_c.fit(trainX_c, trainY_c, epochs=3)
history_d = model_d.fit(trainX_d, trainY_d, epochs=3)
```

```python
score_a = model_a.evaluate(testX_a, testY_a, verbose=0)
score_b = model_b.evaluate(testX_b, testY_b, verbose=0)
score_c = model_c.evaluate(testX_c, testY_c, verbose=0)
score_d = model_d.evaluate(testX_d, testY_d, verbose=0)
print(score_a)
print(score_b)
print(score_c)
print(score_d)
```
